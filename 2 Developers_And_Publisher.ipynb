Game Developers and Publisher

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Dataset = pd.read_csv('vgchartz-2024.csv')
df = Dataset
```

```python
df["developer"].value_counts() .head(10) .plot(kind='bar')
```

```python
df["publisher"].value_counts() .head(10) .plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,2,)
df["developer"].value_counts() .head(10) .plot(kind='bar', ax=ax[0], xlabel="DEVELOPER") 
df["publisher"].value_counts() .head(10) .plot(kind='bar', ax=ax[1], xlabel="PUBLISHER") 

plt.tight_layout()
plt.show()
```
